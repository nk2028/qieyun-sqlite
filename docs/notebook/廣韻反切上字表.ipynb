# 廣韻反切上字表

```python
import pandas
```

```python
import sqlite3
```

```python
conn = sqlite3.connect('../data.sqlite3')
```

```python
pandas.read_sql_query('''SELECT initial,
REPLACE(GROUP_CONCAT(DISTINCT upper_char), ',', '') AS 'upper_chars'
FROM full_guangyun
GROUP BY initial
ORDER BY CASE initial
WHEN '見' THEN 0
WHEN '溪' THEN 1
WHEN '羣' THEN 2
WHEN '疑' THEN 3
WHEN '端' THEN 4
WHEN '透' THEN 5
WHEN '定' THEN 6
WHEN '泥' THEN 7
WHEN '知' THEN 8
WHEN '徹' THEN 9
WHEN '澄' THEN 10
WHEN '孃' THEN 11
WHEN '幫' THEN 12
WHEN '滂' THEN 13
WHEN '並' THEN 14
WHEN '明' THEN 15
WHEN '精' THEN 16
WHEN '清' THEN 17
WHEN '從' THEN 18
WHEN '心' THEN 19
WHEN '邪' THEN 20
WHEN '莊' THEN 21
WHEN '初' THEN 22
WHEN '崇' THEN 23
WHEN '生' THEN 24
WHEN '俟' THEN 25
WHEN '章' THEN 26
WHEN '昌' THEN 27
WHEN '船' THEN 28
WHEN '書' THEN 29
WHEN '常' THEN 30
WHEN '影' THEN 31
WHEN '曉' THEN 32
WHEN '匣' THEN 33
WHEN '云' THEN 34
WHEN '以' THEN 35
WHEN '來' THEN 36
WHEN '日' THEN 37
END, small_rhyme_id;''', conn)
```
