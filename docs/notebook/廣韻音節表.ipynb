```python
from collections import defaultdict
from IPython.display import HTML
import pandas
import sqlite3
```

```python
conn = sqlite3.connect('../data.sqlite3')
```

```python
pandas.options.display.max_rows = None
pandas.options.display.max_columns = None
pandas.options.display.max_colwidth = -1
```

```python
d = defaultdict(dict)
```

```python
for small_rhyme, initial, name in pandas.read_sql_query('''SELECT rhyme || rounding ||
CASE division
WHEN '1' THEN '一'
WHEN '2' THEN '二'
WHEN '3' THEN '三'
ELSE '四'
END AS 'small_rhymes', initial, MIN(name) as 'name'
FROM full_guangyun
GROUP BY small_rhymes, initial
ORDER BY small_rhyme_id;''', conn).itertuples(index=False, name=None):
    d[small_rhyme][initial] = name
```

```python
initials = tuple('見溪羣疑端透定泥知徹澄孃幫滂並明精清從心邪莊初崇生俟章昌船書常影曉匣云以來日')
```

```python
small_rhymes=tuple(d)
```

```python
df = pandas.DataFrame([[d[small_rhyme].get(initial) for initial in initials] for small_rhyme in small_rhymes], index=small_rhymes, columns=initials)
```

```python
for d in df:
    df[d] = df[d].apply(lambda x: f'<a href="https://ytenx.org/zim?dzih={x}&dzyen=1&jtkb=1&jtkd=1&jtdt=1&jtgt=1">{x}</a>' if x else '○')
```
